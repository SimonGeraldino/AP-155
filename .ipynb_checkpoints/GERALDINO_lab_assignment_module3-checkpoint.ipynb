### PROBLEM 1
**Fourier transform of a simple function**

1. Calculate the coefficients in the discrete Fourier transforms of the modulated sine wave $y_n = \sin(\pi n/N) sin(20\pi n/N)$ sampled at $N = 1000$ evenly spaced points using your own user-defined function. *(15 pts.)*

2. Make a plot of their amplitudes similar to the plot shown in Fig. 7.4 of the Newman text. *(15 pts.)*


````python
from numpy import zeros
from cmath import exp,pi

# This defines the Discrete Foruier Transform (DFT) function, referenced from the Newman Textbook
def dft(y):

    # This sets the number of elements in the array y into the value of N, which is the number of points, used
    N = len(y)

    # We create an array, filled with (N//2+1) zeros,
    c = zeros(N//2+1,complex)

    # This for loop returns N/2+1 coefficients for even N and (N+1)/2 coefficients for odd N
    for k in range(N//2+1):

        # Creates multiple instances of n, from the number of values of N
        for n in range(N):

            # This adds the instances of
            c[k] += y[n]*exp(-2j*pi*k*n/N)

    return c
````

````python
import numpy as np

# This is the number of evenly spaced points
N = 1000

# This creates an empty array, to be used later
y = []

# This for loop is used to created multiple instances of y_n from the different values of n and N
for n in np.arange(N):

  # This was the given modulated sine wave
  y_n = np.sin(pi*n/N) * np.sin(20*pi*n/N)

  # We put the values of y_n from the for loop to array y
  y = np.append(y, y_n)
````

````python
import matplotlib.pyplot as plt

# This is used for creating figures and axes
fig, ax = plt.subplots()

# For now, we can create a graph that shows the modulated sine wave given in the problem.
ax.plot(y)
ax.set(title = 'Modulated sine wave of $y_n$ for N = 1000',
       xlabel = 'n',
       ylabel = '$y_n$')

# This is to show the graph
plt.show()
````

````python
# This puts the coefficients created from using the DFT function with the y values
c = dft(y)

# This is used for creating figures and axes
fig,ax = plt.subplots()

# We create a graph that shows the Coefficients from DFT of y
ax.plot(abs(c))
ax.set(title = 'Amplitudes of the Fourier coefficients from DFT of y',
       xlabel = 'k',
       ylabel = '|$c_k$|')

# This is to show the graph
plt.show()
````

````python
# We get the frequency of the primary wave
wave_freq = np.max(abs(c))
print("The frequency of the primary wave is " + str(wave_freq))
````

````python
# We are creating the variable y to store the values from the array dow
y = dow

# This is used for creating figures and axes
fig,ax = plt.subplots()

# For now, we create this graph to show the Daily Closing Price of the Dow Jones Industrial Average (from late 2006 to end of 2010)
ax.plot(y)
ax.set(title = 'Daily Closing Price of Dow Jones Industrial Average (Original Signal)',
       xlabel = 'Day (n)',
       ylabel = 'USD ($y_n$)')

# This is to show the graph
plt.show()
````

````python
# This imports the fft (Fast Fourier Transform) library, with both rfft (Real Fast Fourier Transform) and irfft (Inverse Fast Fourier Transform) from numpy
from numpy import fft

# We are getting the coefficients after RFFT function on y
c = fft.rfft(y)

# This gives this array the values of c
c_10percent = c

# We are creating this variable to find the number of values for the 10 percent of the array
N_10percent = int(len(c) * 0.10)

# This for loop sets up all but 10 percent of the array to zero
for N in np.arange(N_10percent, len(c)):
  c_10percent[N] = 0

# We are using the Inverse Real Fast Fourier Transform using the values from the c_10percent array
dow_filtered = fft.irfft(c_10percent)
````

````python
# This is used for creating figures and axes
fig,ax = plt.subplots()

# This graph is made after filtering the original signal
ax.plot(dow_filtered)
ax.set(title = 'Daily Closing Price of Dow Jones Industrial Average (Filtered Signal)',
       xlabel = 'Day (n)',
       ylabel = 'USD ($y_n$)')

# This is to show the graph
plt.show()
````

````python
# This is used for creating figures and sets up the axes for the "overlapping" graph
fig = plt.figure()
ax1 = fig.add_subplot()

# This is the overlapped graph of the original and the filtered signals of the Daily Closing Price of the Dow Jones Industrial Average
ax1.plot(y, label = 'Original', color = 'red')
ax1.plot(dow_filtered, label = 'Filtered', color = 'blue')
ax1.set(title = 'Daily Closing Prices of Dow Jones Industrial Average',
        xlabel = 'Day (n)',
        ylabel = 'USD ($y_n$)')
ax1.legend(loc='best')

# This is to show the graph
plt.show()
````

The new signal, after the Fourier Transforms, is now filtered and it looks cleaner after getting rid of the "white noise". After looking at this filtered signal, we can say that the first 10% of the array had most of what comprises the "main frequency" and the remaining 90% were probably mostly consisted of the noise.
