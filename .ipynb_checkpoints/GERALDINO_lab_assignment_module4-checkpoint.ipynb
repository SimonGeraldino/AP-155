### PROBLEM 1
**The Lotka-Volterra equations**

_Refer to Exercise 8.2 in the Newman text._ The Lotka-Volterra equations are a mathematical model of predator-prey interactions between biological species. Let two variables $x$ and $y$ be proportional to the size of the populations of two species, traditionally called "rabbits" (the prey) and "foxes" (the predators). You could think of $x$ and $y$ as being the population in thousands, say, so that $x = 2$ means there are 2000 rabbits.

In the Lotka-Volterra model the rabbits reproduce at a rate proportional to their population, but are eaten by the foxes at a rate proportional to both their own population and the population of foxes:

$$\frac{{\rm d}x}{{\rm d}t} = \alpha x - \beta xy,$$

where $\alpha$ and $\beta$ are constants. At the same time the foxes reproduce at a rate proportional to the rate at which they eat rabbits-- because they need food to grow and reproduce-- but also die of old age at a rate proportional to their own population:

$$\frac{{\rm d}y}{{\rm d}t}  = \gamma xy - \delta y,$$

where $\gamma$ and $\delta$ are also constants. *(25 pts.)*

1. Solve these equations using the fourth-order Runge-Kutta method for the case $\alpha = 1, \beta = \gamma = 0.5$, and $\delta = 2$, starting from the initial condition $x = y = 2$.

2. Make a graph showing both $x$ and $y$ as a function of time on the same axes from $t =0$ to $t = 30$. *(25 pts.)*

```python
# Since we are dealing with differential equations with more than one variable,
# we are going to use the code snippet from Example 8.5 found in the Newman textbook

from math import sin
import numpy as np
import matplotlib.pyplot as plt

# These are the values of constants that was given by the problem
alpha = 1
beta = 0.5
gamma = 0.5
delta = 2

# We are setting up f, which is the vector of functions, and is equal to dr/dt
def f(r,t):
    x = r[0]
    y = r[1]
    fx = alpha*x - beta*x*y
    fy = gamma*x*y - delta*y
    return np.array([fx,fy],float)

# These are the bounds for time, with a as the lower bound and b as the upper bound
a = 0.0
b = 30.0

# This is the number of steps
N = 1000

# This is the width of each step
h = (b-a)/N

# This creates an array for time every step of h
tpoints = np.arange(a,b,h)

# This creates an empty array for the x value
xpoints = []

# This creates an empty array for the y value
ypoints = []

# This sets up vector r, with the initial values for x and y
r = np.array([2.0,2.0],float)

# This is the loop for the fourth-order Runge-Kutta method
for t in tpoints:

    # Adds the x value of vector r to array "xpoints"
    xpoints.append(r[0])

    # Adds the y value of vector r to array "ypoints"
    ypoints.append(r[1])

    # These are the k-values
    k1 = h*f(r,t)
    k2 = h*f(r+0.5*k1,t+0.5*h)
    k3 = h*f(r+0.5*k2,t+0.5*h)
    k4 = h*f(r+k3,t+h)

    # This adds the second term of Eq. 8.17e in the Newman textbook and the "initial"
    # vector r to create the "new" vector r
    r += (k1+2*k2+2*k3+k4)/6

# This creates the figures and the axes
fig,ax = plt.subplots()

# This creates the plot for the visual representation for the Lotka-Volterra model
# of the population of rabbits and foxes using the fourth-order Runga-Kutta method
ax.plot(tpoints, xpoints, label = 'Rabbits', color = 'red')
ax.plot(tpoints, ypoints, label = 'Foxes', color = 'blue')

ax.set(title = 'Lotka-Volterra model of the population of rabbits and foxes',
       xlabel = 'Time (t)',
       ylabel = 'Population (thousands)')
ax.legend(loc = 'best')

# Shows the plot
plt.show()
```

The Lotka-Volterra model shows that trend of the prey-predator population over time. In this model, the prey used is the rabbit, and the predator, is the fox.

The trend for both rabbits and foxes were shown to be similar, in a sense that seem to oscillate in the same frequency. Another observation seen is that the trend for foxes is shifted when compared to the rabbit's

### PROBLEM 2

**Nonlinear pendulum**

_Refer to Exercise 8.4 in the Newman text._ Building on the results from Example 8.6, calculate the motion of a nonlinear pendulum as follows.

1. Solve the two first-order equations, Eqs. (8.45) and (8.46), using the fourth-order Runge-Kutta method for a pendulum with a 10 cm arm. Use your program to calculate the angle $\theta$ of displacement for several periods of the pendulum when it is released from a standstill at $\theta = 179^\circ$ from the vertical. *(25 pts.)*

2. Make a graph of $\theta$ as a function of time. *(25 pts.)*


```python
# From Exercise 8.6 in the Newman textbook

# These are the constants used for the question
g = 9.81    # The acceleration due to gravity
l = 0.1     # The length of the pendulum arm

# We are setting up f, which is the vector of functions, and is equal to dr/dt
def f(r,t):
    theta = r[0]
    omega = r[1]
    ftheta = omega
    fomega = -(g/l)*sin(theta)
    return np.array([ftheta,fomega],float)

# These are the bounds for time, with a as the lower bound and b as the upper bound
a = 0.0
b = 10.0

# This is the number of steps
N = 1000

# This is the width of each step
h = (b-a)/N

# This creates an array for time every step of h
tpoints = np.arange(a,b,h)

# This creates an empty array for the theta value
thetapoints = []

# This creates an empty array for the omega value
omegapoints = []

# This sets up vector r, with the initial values for theta and omega
r = np.array([np.radians(179),0],float)

# This is the loop for the fourth-order Runge-Kutta method
for t in tpoints:

    # Adds the theta value of vector r to array "thetapoints"
    thetapoints.append(r[0])

    # Adds the omega value of vector r to array "omegapoints"
    omegapoints.append(r[1])

    # These are the k-values
    k1 = h*f(r,t)
    k2 = h*f(r+0.5*k1,t+0.5*h)
    k3 = h*f(r+0.5*k2,t+0.5*h)
    k4 = h*f(r+k3,t+h)

    # This adds the second term of Eq. 8.17e in the Newman textbook and the "initial"
    # vector r to create the "new" vector r
    r += (k1+2*k2+2*k3+k4)/6

# This creates the figures and the axes
fig,ax = plt.subplots()

# This creates the plot for the visual representation for the Nonlinear pendulum
# with varying angle theta over time, using the fourth-order Runga-Kutta method
ax.plot(tpoints, np.degrees(thetapoints))
ax.set(title = 'Nonlinear pendulum (using fourth-order Runga-Ketta Method)',
       xlabel = 'Time (t)',
       ylabel = 'Angle (theta)',
       xlim = [0,10])

# Shows the plot
plt.show()
```

This graphical representation of the movement of the nonlinear pendulum, with respect to its angle from the vertical over time, shows an oscillatory trend.
